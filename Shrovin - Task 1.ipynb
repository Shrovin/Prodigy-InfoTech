## Extraction of top ten countries with respect to total population

```python
#Filter data for total population
total_population_data=df[df['Series Code']=='SP.POP.TOTL']

#sort data based on total population for 2022
total_population_sorted = total_population_data.sort_values(by="2022",ascending=False)

#Get the top ten countries with highest total number of population for 2022
total_top_ten_countries=total_population_sorted.head(10)
print("Top ten countries of total population\n")
print(total_top_ten_countries[['Country Code']])
```

# Stacked Bar plot

## Top 10 countries with male and female population

```python
# merge male and female data on the country code
merge_data = pd.merge(male_population_data,female_population_data,on="Country Code",suffixes=("_male","_female"))
```

```python
#merged data
# calculate the population of each country (male+female)
merge_data["Total population"] = merge_data["2022_male"]+merge_data["2022_female"]
```

```python
merge_data.head(10)
```

```python
#sort data based on total population in descending order
sorted_data = merge_data.sort_values(by="Total population", ascending = False)
```

## Select the top 10 countries with highest total population

```python
top_10_countries = sorted_data.head(10)
```

```python
#create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(data=top_10_countries,x="Country Code",y="2022_female",color="blue",label="female population")
sns.barplot(data=top_10_countries,x="Country Code",y="2022_male",bottom=top_10_countries["2022_female"],color="red",label="male population")
plt.xlabel("Country",fontsize=15)
plt.ylabel("Population",fontsize=15)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

## Bottom 10 countries with male and female population (2022)

```python
#Select the bottom 10 countries with highest male and female population
bottom_10_countries = sorted_data.tail(10)
```

```python
bottom_10_countries
```

```python
#create the stacked bar plot
plt.figure(figsize=(12,6))
sns.barplot(data=bottom_10_countries,x="Country Code",y="2022_female",color="yellow",label="female population")
sns.barplot(data=bottom_10_countries,x="Country Code",y="2022_male",bottom=bottom_10_countries["2022_female"],color="green",label="male population")
plt.xlabel("Country",fontsize=15)
plt.ylabel("Population",fontsize=15)
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
